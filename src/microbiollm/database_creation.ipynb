```python
import pandas as pd

from ete3 import NCBITaxa
```

## Helpers

```python
def tax_id_translator(tax_id: str):
    ncbi = NCBITaxa()
    name = ncbi.get_taxid_translator([tax_id])
    return name[tax_id]


def get_taxonomy(tax_id: str, rank: str):
    ncbi = NCBITaxa()
    try:
        lineage = ncbi.get_lineage(tax_id)
        rank_id = [kegit@github.com:AndySECP/MicroBioLLM.gity for key, value in ncbi.get_rank(lineage).items() if value == rank]
        return tax_id_translator(rank_id[0])
    except:
        return "NA"
```

## Assembly summary

```python
assembly_summary = pd.read_csv("assembly_summary_refseq.txt", skiprows=1, sep= "\t")
```

```python
complete_genome = assembly_summary[assembly_summary["assembly_level"] == "Complete Genome"]
```

```python
complete_genome['superkingdom'] = complete_genome['taxid'].apply(lambda x: get_taxonomy(str(x), "superkingdom"))
```

```python
bacteria = complete_genome[complete_genome["superkingdom"] == "Bacteria"]
```

```python
bacteria['genus'] = complete_genome['taxid'].apply(lambda x: get_taxonomy(str(x), "genus"))
```

```python
bacteria['family'] = complete_genome['taxid'].apply(lambda x: get_taxonomy(str(x), "family"))
```

```python
families = ['Enterobacteriaceae', 'Bacillaceae', 'Pseudomonadaceae', 'Staphylococcaceae', 'Streptococcaceae']
```

```python
def select_bacterial_set(families: list, bacteria_df): 
    selected_genomes = pd.DataFrame()

    for family in families:
        genomes_family = bacteria_df[bacteria_df['family'] == family].sample(n=100, random_state=42)  # Adjust random_state as needed for reproducibility
        selected_genomes = pd.concat([selected_genomes, genomes_family])
    return selected_genomes
    
```

```python
selected_genomes = select_bacterial_set(families, bacteria)
```

```python
list(selected_genomes[["ftp_path", "#assembly_accession"]]['#assembly_accession'])
```

```python
to_download = []

for species in list(selected_genomes['ftp_path']): 
    to_download.append(f'{species}/{species.split("/")[-1]}_genomic.fna.gz')
```

```python
def save_list_to_file(lst, filename):
    with open(filename, 'w') as f:
        for item in lst:
            f.write(str(item) + '\n')
    print(f"List has been saved to {filename}")
```

```python
save_list_to_file(to_download, 'to_download.txt')
```

```python
to_download.to_csv('download_genomes.txt')
```

### Generate fastq files

```python
import random
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord

def generate_microbiome_sample(genome_files, composition, read_length, num_reads):
    microbiome_reads = []

    for genome_file, abundance in composition.items():
        # Load the genome
        sequences = SeqIO.parse(genome_file, "fasta")
        genome_reads = []
        
        # Generate reads from the genome
        for sequence in sequences:
            seq_length = len(sequence.seq)
            if seq_length >= read_length:
                for _ in range(int(abundance * num_reads)):
                    start_position = random.randint(0, seq_length - read_length)
                    read_seq = sequence.seq[start_position:start_position+read_length]
                    read_record = SeqRecord(read_seq, id=f"{sequence.id}_read", description=f"Generated read from {sequence.id}")
                    genome_reads.append(read_record)

        # Randomly shuffle reads to avoid sequence bias
        random.shuffle(genome_reads)

        # Add reads to microbiome sample
        microbiome_reads.extend(genome_reads[:int(abundance * num_reads)])

    # Randomly shuffle all the selected reads
    random.shuffle(microbiome_reads)

    return microbiome_reads

# Example usage:
genome_files = {
    'genome1.fasta': 0.2,  # Path to the genome file and its abundance in the composition
    'genome2.fasta': 0.3,
    # Add more genome files and their abundances as needed
}

composition = {
    'genome1.fasta': 0.5,  # Abundance of each genome in the sample
    'genome2.fasta': 0.5,
    # Add more genomes and their abundances as needed
}

read_length = 300  # Length of each read
num_reads = 1000  # Total number of reads in the synthetic microbiome sample

microbiome_reads = generate_microbiome_sample(genome_files, composition, read_length, num_reads)

# Now you have a list of SeqRecord objects containing the synthetic microbiome reads
# You can write them to a new fasta file using SeqIO.write
SeqIO.write(microbiome_reads, "synthetic_microbiome.fasta", "fasta")
```
